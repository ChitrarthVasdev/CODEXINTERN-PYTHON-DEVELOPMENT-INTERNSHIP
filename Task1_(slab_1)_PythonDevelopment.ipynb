```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load your uploaded dataset
df = pd.read_csv('/content/advertising.csv')

# Step 3: Display dataset info
print("ðŸ“Š First 5 rows of the dataset:")
print(df.head())

print("\nðŸ“ˆ Dataset Summary:")
print(df.describe())

print("\nðŸ”¢ Columns:")
print(df.columns.tolist())
```

```python
# Choose a numeric column for average
average_sales = df['Sales'].mean()
print(f"\nâœ… Average Sales: {average_sales:.2f}")
```

```python
# Average Spend per Channel
df[['TV', 'Radio', 'Newspaper']].mean().plot(kind='bar', color=['red', 'blue', 'green'])
plt.title("Average Advertising Spend by Channel")
plt.ylabel("Average Spend")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# TV Spend vs Sales
sns.scatterplot(data=df, x='TV', y='Sales', color='purple')
plt.title("TV Spend vs Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
